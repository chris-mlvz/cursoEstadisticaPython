```python
import pandas as pd 
# import numpy as np 
import matplotlib.pyplot as plt 
# %matplotlib inline
```

```python
from scipy.stats import expon
from scipy.stats import uniform
```

```python
poblacion = pd.DataFrame()
poblacion['numbers'] = expon.rvs(40, size=100000)
```

```python
poblacion['numbers'].hist(bins=100)
```

```python
tamano = 2000
muestras_promedio = [poblacion.sample(n=100).mean()[0] for i in range(tamano)]
```

```python
fig, ax = plt.subplots()
ax.hist(muestras_promedio, bins=50, alpha=0.5)
ax.set_xlabel('Valor promedio')
ax.set_ylabel('Frecuencia')
ax.grid()
```
