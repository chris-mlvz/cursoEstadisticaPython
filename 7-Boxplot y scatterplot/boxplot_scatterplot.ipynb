```python
import numpy as np 
import pandas as pd 
import scipy as sp 

import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("../data/bicicletas-compartidas.csv")
df.columns
```

```python
y = df['bicis-compartidas']

fig, ax = plt.subplots()
ax.boxplot(x=y)
```

```python
sns.boxplot(x='cuartil-ano', y='bicis-compartidas', data=df)
```

```python
fig, ax = plt.subplots()
ax.scatter(x='viento', y='bicis-compartidas', data=df, alpha=0.1)
ax.set_title('Distribucin conjunta de Viento y Bicis-Compartidas')
ax.set_xlabel('Viento')
ax.set_ylabel('Bicis-compartidas')
```
