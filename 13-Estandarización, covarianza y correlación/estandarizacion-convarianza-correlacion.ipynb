```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline
```

```python
df = pd.read_csv('../data/iris-data.csv', index_col = 0)
```

```python
df.columns
```

```python
df.tipo_flor.value_counts()
```

```python
y = df['lar.petalo']
```

```python
fig, ax = plt.subplots()
ax.set_title('Variable original')
ax.hist(y, bins=30)
ax.axvline(x = np.mean(y), c='k', label='media', linestyle = '--')
ax.axvline(x = np.mean(y) + np.std(y), c='r', label='Desviación estandar +', linestyle = '--')
ax.axvline(x = np.mean(y) - np.std(y), c='b', label='Desviación estandar -', linestyle = '--')
ax.legend()
```

```python
fig, ax = plt.subplots()
y_estandar = (y - np.mean(y))/np.std(y)
ax.set_title('Variable original')
ax.hist(y_estandar, bins=30)
ax.axvline(x = np.mean(y_estandar), c='k', label='media', linestyle = '--')
ax.axvline(x = np.mean(y_estandar) + np.std(y_estandar), c='r', label='Desviación estandar +', linestyle = '--')
ax.axvline(x = np.mean(y_estandar) - np.std(y_estandar), c='b', label='Desviación estandar -', linestyle = '--')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(df['lar.petalo'], df['lar.sepalo'], alpha=0.7)
ax.set_xlabel('lar.petalo')
ax.set_ylabel('lar.sepalo')
ax.autoscale()
```
