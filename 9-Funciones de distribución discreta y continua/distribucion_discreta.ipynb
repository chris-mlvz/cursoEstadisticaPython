```python
import numpy as np 
import pandas as pd 
import scipy as sp 

import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
```

```python
from scipy.stats import binom

p = 0.3
n = 8

x = np.arange(n+1)
y = tuple(binom.pmf(i, p=p, n=n) for i in x)

fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('probabilidad discreta')

np.sum(y)
```

```python
media = np.average(x, weights=y) #sum(x*y)
varianza = np.average((x-media)**2, weights=y)
media, varianza
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('probabilidad discreta')
ax.axvline(x=media, c='r', linestyle='--', label='Valor Esperado')
ax.axvline(x=media + 3*np.sqrt(varianza), c='r', linestyle='--', label='Desviacin Estandar')
ax.legend()
```
